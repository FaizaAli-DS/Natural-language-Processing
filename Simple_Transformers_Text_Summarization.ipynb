```python
from transformers import pipeline
import pandas as pd
```

Lets sets up a text summarization pipeline using the Hugging Face Transformers library.

```python
text_summarizer = pipeline("summarization")
```

```python
input_text =""" A transformer neural network can take an input sentence in the form of a sequence of vectors,
 converts it into a vector called an encoding, and then decodes it back into another sequence.
  An important part of the transformer is the attention mechanism,
  which allows the network to focus on specific parts of the input when decoding the output.
   This results in a more efficient and accurate translation of the input sentence.
   As we know RNN, GRU and LSTM are sequential models that have some limitations,
   they cannot work with large sequences. LSTM have some memory concept but still,
    it is not able to recall the whole input.
   That’s why transformers come into the picture they have infinite recall of input using the self-attention technique."""
```

```python
summary = text_summarizer(input_text, max_length =100, clean_up_tokenization_spaces = True )
print(summary[0]['summary_text'])
#summary variable is a list of dictionaries lets print 1st one
```
